# 신용카드 매출액 월별 분석

```python
# 대전광역시 신용카드 매출액 데이터
신용카드 = pd.read_csv('data/대전광역시_자치구별_신용카드_매출액.csv', encoding='cp949')
신용카드
```

```python
신용카드[신용카드.기준년도 == '2019년'].업종.unique()
```

```python
신용카드_19 = 신용카드[신용카드.기준년도 == '2019년']
신용카드_19
```

```python
신용카드_19 = 신용카드_19.groupby('업종').agg({
                          '구':','.join,
                          '3월(억원)':'first',
                          '4월(억원)':'first',
                          '5월(억원)':'first',
                          '6월(억원)':'first',
                          '7월(억원)':'first'})
신용카드_19 = 신용카드_19.drop(['구'], axis=1).fillna(0).T
신용카드_19.index = ['3월', '4월', '5월', '6월', '7월']
신용카드_19
```

```python
신용카드_20 = 신용카드[신용카드.기준년도 == '2020년']
신용카드_20 = 신용카드_20.groupby('업종').agg({
                          '구':','.join,
                          '3월(억원)':'first',
                          '4월(억원)':'first',
                          '5월(억원)':'first',
                          '6월(억원)':'first',
                          '7월(억원)':'first'})
신용카드_20 = 신용카드_20.drop(['구'], axis=1).fillna(0).T
신용카드_20.index = ['3월', '4월', '5월', '6월', '7월']
신용카드_20
```

```python
신용카드_19.plot( figsize=(10,10), title='2019년 신용카드 매출')
신용카드_20.plot( figsize=(10,10), title='2020년 신용카드 매출')
```

# 신용카드 매출액 구별 분석

```python
신용카드[신용카드.기준년도 == '2019년'].구.unique()
```

```python
구별카드_19 = 신용카드[신용카드.기준년도 == '2019년']
구별카드_19 = 구별카드_19.groupby('구').agg({
                          '3월(억원)':'first',
                          '4월(억원)':'first',
                          '5월(억원)':'first',
                          '6월(억원)':'first',
                          '7월(억원)':'first'})
구별카드_19 = 구별카드_19.sum(axis=1)
구별카드_19 = 구별카드_19.reset_index()
구별카드_19.columns = ['구', '매출']
구별카드_19
```

```python
구별카드_20 = 신용카드[신용카드.기준년도 == '2020년']
구별카드_20 = 구별카드_20.groupby('구').agg({
                          '3월(억원)':'first',
                          '4월(억원)':'first',
                          '5월(억원)':'first',
                          '6월(억원)':'first',
                          '7월(억원)':'first'})
구별카드_20 = 구별카드_20.sum(axis=1)
구별카드_20 = 구별카드_20.reset_index()
구별카드_20.columns = ['구', '매출']
구별카드_20
```

```python
import folium
map = folium.Map(location = [36.350594937132485, 127.38477578238654], # 대전시청
                 zoom_start = 11, # 값이 커질수록 디테일
                 zoom_control = False, # zoom in/out 버튼 표시 X
                 control_scale = True, # 스케일 컨트롤 버튼 표시 O
                 tiles = 'https://tiles.stadiamaps.com/tiles/outdoors/{z}/{x}/{y}{r}.png',
                 attr = '&copy; <a href="https://stadiamaps.com/">Stadia Maps</a>, &copy; <a href="https://openmaptiles.org/">OpenMapTiles</a> &copy; <a href="http://openstreetmap.org">OpenStreetMap</a> contributors')
map
```

```python
# GeoJson으로 행정구역경계 표시
import json
with open('data/Daejeon_geo.json', encoding='utf-8') as file:
    geo = json.loads(file.read())
    file.close()
```

```python
choro = folium.Choropleth(geo_data = geo,
                          name = 'choropleth',
                          data = 구별카드_19,
                          columns = ['구', '매출'],
                          key_on = 'feature.properties.sggnm',
                          legend_name = '2019년 구별 카드매출',
                          fill_color = 'BuPu'
                         ).add_to(map)
choro.geojson.add_child(folium.features.GeoJsonTooltip(['temp']))
map
```

```python
choro = folium.Choropleth(geo_data = geo,
                          name = 'choropleth',
                          data = 구별카드_20,
                          columns = ['구', '매출'],
                          key_on = 'feature.properties.sggnm',
                          legend_name = '2020년 구별 카드매출',
                          fill_color = 'BuPu'
                         ).add_to(map)
choro.geojson.add_child(folium.features.GeoJsonTooltip(['temp']))
map
```

# 코로나19 데이터와의 비교

```python
covid = pd.read_excel('data/owid-covid-data.xlsx')
covid['date'] =  pd.to_datetime(covid['date'], format='%Y-%m-%d')
covid['year'] = covid['date'].apply(lambda x: x.strftime('%Y'))
covid['month'] = covid['date'].apply(lambda x: x.strftime('%m'))
covid['day'] = covid['date'].apply(lambda x: x.strftime('%d'))
covid['weekNumber'] = covid['date'].apply(lambda x: x.strftime('%U'))
covid['weekDay'] = covid['date'].apply(lambda x: x.strftime('%a'))
covid = covid.set_index(['date']).sort_index()
covid = covid[covid.location == 'South Korea']
```

```python
covid = covid[covid.month.isin(['03','04','05','06','07'])]
covid
```

```python
covid.columns
```

```python
covid = covid.drop(['iso_code', 'continent', 'location', 'new_cases_smoothed',
       'new_deaths_smoothed', 'total_cases_per_million',
       'new_cases_per_million', 'new_cases_smoothed_per_million',
       'total_deaths_per_million', 'new_deaths_per_million',
       'new_deaths_smoothed_per_million', 'reproduction_rate', 'icu_patients',
       'icu_patients_per_million', 'hosp_patients',
       'hosp_patients_per_million', 'weekly_icu_admissions',
       'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
       'weekly_hosp_admissions_per_million', 'new_tests', 'total_tests',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'positive_rate', 'tests_per_case', 'tests_units', 'total_vaccinations',
       'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations',
       'new_vaccinations_smoothed', 'total_vaccinations_per_hundred',
       'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred',
       'new_vaccinations_smoothed_per_million', 'stringency_index',
       'population', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand',
       'life_expectancy', 'human_development_index', 'year', 'month', 'day',
       'weekNumber', 'weekDay'], axis=1).fillna(0)
```

```python
covid
```

```python
covid.plot( figsize=(10,6) )
```

```python
data1 = covid.drop(['new_cases','total_deaths','new_deaths'], axis=1)
data1 = data1.loc[['2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30', '2020-07-31']]
data1.index = ['3월', '4월', '5월', '6월', '7월']
data1
```

```python
data2 = covid.drop(['total_cases','total_deaths','new_deaths'], axis=1)
data2 = data2.loc[['2020-03-31', '2020-04-30', '2020-05-31', '2020-06-30', '2020-07-31']]
data2.index = ['3월', '4월', '5월', '6월', '7월']
data2
```

```python
plt.rcParams['figure.figsize'] = (10,8)
plt.rcParams['font.size'] = 15

fig, ax1 = plt.subplots()
ax1.set_ylabel('누적 확진자수(명)')
ax1.plot(data1.index, data1.values, color='blue', marker='o', markersize=9)
ax1.legend('누', loc='upper right', fontsize=14)

ax2 = ax1.twinx()
ax2.set_ylabel('신규 확진자수(명)')
ax2.plot(data2.index, data2.values, color='orange', marker='*', markersize=12)
ax2.legend('신', loc='upper left', fontsize=14)

plt.plot(신용카드_20.index, 신용카드_20.values, alpha=0.4)
plt.show()
```
