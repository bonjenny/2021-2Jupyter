### 2) read_html() - html 문서 내 테이블 데이터 읽어오기

- 시가총액 테이블도 정보로 보고 읽어들여왔기 때문에
- 두번째 테이블이 우리가 원하는 정보

```python
data = pd.read_html('https://finance.naver.com/sise/sise_market_sum.naver', encoding = 'cp949')
```

```python
df = data[1]
```

```python
# NaN값 제거하기
# 종목명의 값이 NaN이 아니라면 True, 맞다면 False
df[df.종목명.notnull()]
```

- 50개의 페이지 중 10개의 페이지만 읽어들여와 하나의 데이터로 합치기

```python
# 시가총액이 높은 상위 500개 기업 데이터 읽어오기

코스피500 = []
url = 'https://finance.naver.com/sise/sise_market_sum.naver?&page='
for num in range(1,11):
    full_url = url + str(num)
    data = pd.read_html(full_url, encoding = 'cp949')
    df = data[1]
    df = df[df.종목명.notnull()]
    코스피500.append(df)
```

```python
# 항목당 50개씩 총 500개의 데이터가 저장됨
코스피500[0]
```

```python
data = pd.concat(코스피500)
```

```python
data.to_html('data/코스피500데이터.html', encoding = 'cp949')
```

### 3) Selenium 라이브러리 활용  

* selenium 설치: conda install selenium
* chromedriver 다운로드 : https://sites.google.com/a/chromium.org/chromedriver/downloads

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('chromedriver')
```

```python
driver.implicitly_wait(3)
```

```python
driver.get('https://finance.naver.com/item/sise_day.nhn?code=005930')
```

```python
html = driver.page_source
```

```python
data = pd.read_html(html, encoding = 'cp949')
```

```python
삼성_일별 = data[0][data[0].날짜.notnull()]
```

```python
# 실습. 모든 기간의 삼성전자 일별 데이터를 스크래핑해서
# 하나의 데이터프레임으로 저장

일별주가 = []
삼성전자url = 'https://finance.naver.com/item/sise_day.nhn?code=005930&page='
for i in range(1,200):
    full_url = 삼성전자url + str(i)
    driver.get(full_url)
    html = driver.page_source
    data = pd.read_html(html, encoding = 'cp949')
    삼성_일별 = data[0][data[0].날짜.notnull()]
    일별주가.append(삼성_일별)
```

```python
삼성전자_일별_최종데이터 = pd.concat(일별주가)
```

```python
삼성전자_일별_최종데이터
```

```python
삼성전자_일별_최종데이터.to_excel('data/삼성전자_일별주가.xlsx', encoding = 'cp949')
```

## 4. 데이터베이스에서 읽기
### 파이썬에 내장(embedded)된 sqlite 사용

#### 데이터베이스란?
- 데이터베이스를 관리하기 위해서 사용하는 프로그램, DBMS
> MySQL, Oracle, PostgreSQL

```python
import sqlite3
```

```python
#테이블 생성 SQL 정의
테이블생성 = 'CREATE TABLE locations \
                (a varchar[20], b varchar[20], c integer)'
```

```python
#sqlite 접속
conn = sqlite3.connect('temp.db')
```

```python
#테이블 생성 SQL 수행 
conn.execute(테이블생성)
```

```python
# commit() 수행. - SQL 수행 후, commit을 해야만, 데이터베이스에 SQL 구문이 적용됨 
conn.commit()
```

```python
#생성된 테이블에 저장할 데이터 정의
data = [['서울','강남', 15]
       ,['서울','강동', 10]
       ,['서울','동대문', 7]
       ,['서울','양천', 12]
       ,['성남','분당', 12]
       ,['수원','영통', 15]
       ,['성남','중원', 11]
       ,['수원','팔달', 14]
       ]
```

```python
# 데이터 적재 SQL 정의 
삽입 = 'insert into locations values (?,?,?)'
```

```python
#데이터 적재 SQL 수행 
conn.executemany(삽입, data)
```

```python
# commit() 수행 
conn.commit()
```

```python
#Select SQL 실행1
pd.read_sql('select * from locations', conn)
```

```python
#Select SQL 실행2
pd.read_sql('select a, sum(c) from locations group by a', conn)
```

```python
# 데이터베이스의 테이블 크기가 큰 경우에는 cursor 사용하기
cursor = conn.execute('select * from locations')
```

```python
cursor
```

- cursor.fetchmany(cnt): cnt 개수만큼 데이터를 읽어옴
- 다시 실행하면 그 다음 데이터 읽어옴
- 데이터를 끝까지 다 읽으면 비어 있는 리스트가 반환됩니다

```python
cursor.fetchmany(3)
```

- cursor나 데이터베이스의 conn(커넥션) 같은 경우
- close()를 해주는게 기본적인 자원 관리를 위해 필요함

```python
cursor.close()
```
